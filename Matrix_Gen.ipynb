```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
def PolyDecDiagonal(R,n,p):
    D_0 = np.ones(R)
    D_1 = np.arange(2, n-R+2)**(-p)
    D = np.concatenate((D_0, D_1))
    return D

def ExpDecDiagonal(R,n,q):
    D_0 = np.ones(R)    
    D_1 = np.exp(-q*np.arange(2, n-R+2))
    D = np.concatenate((D_0, D_1))
    return D
```

```python
n = int(2**14)
Rs = [5,10,20]
p_poly = [0.5, 1, 2]
q_exp = [.1, .25, 1]
```

```python
with open('A_Poly_test.pkl', 'wb') as f:
    pickle.dump(PolyDecDiagonal(5,1024,0.5), f)

with open('A_Exp_test.pkl', 'wb') as f:
    pickle.dump(ExpDecDiagonal(5,1024,.1), f)
```

```python
with open('A_Poly_test.pkl', 'rb') as f:
    matrix = pickle.load(f) 
plt.plot(np.arange(512), matrix)
plt.title('Polynomial Decay, R=5, p=0.5')
plt.show()
```
